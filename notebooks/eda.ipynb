```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/LENOVO/OneDrive/Desktop/DMW project/data/processed/ingested.csv")
print(df.head())
```

```python
df.shape
df.describe()
df.info()
```

```python
# spedning distribution

import matplotlib.pyplot as plt

df['amount'].hist(bins=30)
plt.title("spending distribution")
plt.show()
```

```python
# Category wise spending

df.groupby("category")['amount'].sum().sort_values().plot(kind='barh')
plt.title("spending by category")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure columns exist
df['month'] = pd.to_datetime(df['date']).dt.strftime('%b')  # e.g., Jan, Feb...

df.groupby('month')['amount'].sum().plot(kind='bar')
plt.title("Monthly Spending Trend")
plt.tight_layout()
plt.show()
```

```python
# user wise behaviour 

df.groupby('user_id')['amount'].sum().sort_values(ascending=False).head(10)
```

```python
# payment type analysis 

df.groupby('payment_type')['amount'].sum().plot(kind='bar')
plt.title("Payment Type Usage")
plt.show()
```

```python
# outlier boxplot 

sns.boxplot(x=df['amount'])
plt.title("Outlier Detection - Amount")
plt.show()
```

```python
user_spend = df.groupby('user_id')['amount'].sum()

sns.histplot(user_spend, bins=20)
plt.title("User Total Spending Distribution")
plt.show()
```
